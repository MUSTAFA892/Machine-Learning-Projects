```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
# Label Encoding
data.replace({'Loan_Status':{'N':0, 'Y':1}}, inplace=True)
```

```python
data.head()
```

```python
data['Dependents'].value_counts()
```

```python
data = data.replace(to_replace='3+', value=4)
```

```python
data['Dependents'].value_counts()
```

```python
data['Property_Area'].value_counts()
```

```python
data.replace({'Married': {'No':0, 'Yes':1}, 'Gender': {'Male':1, 'Female':0}, 'Self_Employed': {'No':0, 'Yes':1}, 'Property_Area':{'Rural':0, 'Semiurban':1,'Urban':2}, 'Education': {'Graduate':0,'Not Graduate':1}}, inplace=True)
```

```python
data.head()
```

```python
X = data.drop(columns=['Loan_ID','Loan_Status'], axis=1)
Y = data['Loan_Status']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_accuracy_score = accuracy_score(X_train_prediction,Y_train)
```

```python
print("The accuracy of training data:", training_accuracy_score)
```

```python
X_test_prediction = classifier.predict(X_test)
testing_accuracy_score = accuracy_score(X_test_prediction,Y_test)
```

```python
print("The accuracy of testing data:", testing_accuracy_score)
```
