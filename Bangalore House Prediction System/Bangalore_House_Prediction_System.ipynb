```python
#To handle this type of index range values 1133 - 1384
def ConvertRange(x):
  temp = x.split('-')
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

###Other detection and removal

```python
data.describe()
# Here it shows that in min some gives 1 sqft so it can be considered as the outliers
```

```python
(data['total_sqft'] / data['bhk']).describe()
```

```python
data = data[((data['total_sqft'] / data['bhk']) >= 300)]
data.describe()
```

```python
def remove_outliers_sqft(df):
  df_output = pd.DataFrame()
  for key , subdf in df.groupby('location'):
    m = np.mean(subdf.price_per_sqft)
    st = np.std(subdf.price_per_sqft)
    gen_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output,gen_df],ignore_index = True)
  return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location , location_df in df.groupby('location'):
    bhk_stats = {}
    for bhk , bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk] = {
          'mean' : np.mean(bhk_df.price_per_sqft),
          'std' : np.std(bhk_df.price_per_sqft),
          'count' : bhk_df.shape[0]
      }
    for bhk , bhk_df in location_df.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis = 'index')
```

```python
data = bhk_outlier_remover(data)
```

```python
X = data.drop(columns = ['price'])
Y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

###Applying Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler , lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(Y_test,y_pred_lr)
```

###Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(Y_test,y_pred_lasso)
```

###Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(Y_test, y_pred_ridge)
```

```python
print("Linear Regression:", r2_score(Y_test, y_pred_lr))
print("Lasso Regression:", r2_score(Y_test,y_pred_lasso))
print("Ridge Regression:", r2_score(Y_test,y_pred_ridge))
```
